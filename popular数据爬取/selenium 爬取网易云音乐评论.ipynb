```python
from selenium import webdriver
from lxml import etree
import time

#实例化一个浏览器对象(传入浏览器的驱动程序)
driver_path = r"E:\chromedriver\chromedriver.exe"    #我将chrome驱动安装的路径
driver = webdriver.Chrome(executable_path = driver_path)

#让浏览器发起一个指定url对应的请求
driver.get("https://music.163.com/#/album?id=35069014")

#page_source获取浏览器当前页面的页面源码数据
page_text = driver.page_source

# 先将第一页的源码数据存放到列表中，然后将后续获得的页面源码数据append到列表中
all_page_text = [page_text]
driver.switch_to_frame("contentFrame")

#执行一组js程序  （例如：让屏幕滚动）
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
time.sleep(1)

for i in range(6):
    
    # 进行下一页按钮的定位且对其进行点击
    next_tag = driver.find_element_by_partial_link_text("下一页")
    next_tag.click()
    time.sleep(2)
    all_page_text.append(driver.page_source)
    
    for page_text in all_page_text:
        print(page_text)
        
#         tree = etree.HTML(page_text)
#         alists = tree.xpath('//*[@id="auto-id-JtBIWrFGdpuSCTmG"]')
#         for alist in alists:
#             a_list = alist.xpath('.//div[@class = "itm"]')
#             print(a_list)
```
