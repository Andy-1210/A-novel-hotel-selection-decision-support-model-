```python
from selenium import webdriver
import csv
import time

import pandas as pd
address="./马蜂窝数据提取.xlsx"

result = []
def excel_one_line_to_list():
    df = pd.read_excel(address, usecols=[11],
                       names=None)  # 读取第一列，不要列名
    df_li = df.values.tolist()

    for s_li in df_li:
        result.append(s_li[0])
    return result

if __name__ == '__main__':
    excel_one_line_to_list()

# 新建文件保存
csvf = open('./5001-全部详情数据测试.csv','a+',encoding='gbk',newline ='')
writer = csv.writer(csvf)
writer.writerow(('a1','a6','a2','a4','a3'))


#实例化一个浏览器对象(传入浏览器的驱动程序)
driver_path = r"E:\chromedriver\chromedriver.exe"    #我将chrome驱动安装的路径
driver = webdriver.Chrome(executable_path = driver_path)


for url in result[5000:]:
    #让浏览器发起一个指定url对应的请求
    driver.get(url)

    time.sleep(5)
    try:
        a1 = driver.find_element_by_xpath('//*[@id="_j_path_number"]/b[3]').text
        a2 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/div[4]/span[1]/a').text
        a3 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/div[7]/div[1]/strong/a').text
        a4 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/div[7]/div[2]/strong/a').text
        a5 = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[2]/div/ul/li[2]/a').click()
        time.sleep(6)
        a6 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/div/div[1]/div/div[3]/strong').text
        try:
            writer.writerow((a1,a6,a2,a4,a3))
            print(a1,a6,a2,a4,a3)

        except:
            pass
    except:
        a1 = 0
        a2 = 0
        a3 = 0
        a4 = 0
        a5 = 0
        writer.writerow((a1,a6,a2,a4,a3))
csvf.close()

```
