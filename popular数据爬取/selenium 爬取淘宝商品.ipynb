#### 获取首页商品的所有详细链接

- 目标是获取商品的名称 价格 销量 评论数

```python
# 获取所有商品的链接
detail_url = driver.find_elements_by_xpath('//div[@class="row row-2 title"]/a[@href]')
```

#### 对首页的第一个商品详情页进行数据爬取

```python
detail_url[0].click()
```

```python
# 操作对象 切换到打开的页面
driver.switch_to.window(driver.window_handles[1])
```

```python
# 获取商品名称
title = driver.find_element_by_xpath('//h1[@data-spm="1000983"]').text
price = driver.find_element_by_xpath('//span[@class="tm-price"]').text
sale_count = driver.find_element_by_xpath('//span[@class="tm-count"]').text
comment_count = driver.find_element_by_xpath('//span[@class="tm-count"]').text
print(title,price,sale_count,comment_count)

# 关闭当前页面
driver.close()

# 操作对象 切换到打开的页面
driver.switch_to.window(driver.window_handles[0])
```

#### 对首页全部商品的详情页数据爬取

```python
titles = []
prices = []
sale_counts = []
comment_counts = []

for i in detail_url:
    i.click()
    # 操作对象 切换到打开的页面
    driver.switch_to.window(driver.window_handles[1])
    
    try:
        # 获取商品名称
        title = driver.find_element_by_xpath('//h1[@data-spm="1000983"]').text
        titles.append(title)
        price = driver.find_element_by_xpath('//span[@class="tm-price"]').text
        prices.append(price)
        sale_count = driver.find_element_by_xpath('//span[@class="tm-count"]').text
        sale_counts.append(sale_count)
        comment_count = driver.find_element_by_xpath('//span[@class="tm-count"]').text
        comment_counts.append(comment_count)
        print(title,price,sale_count,comment_count)

        # 关闭当前页面
        driver.close()

        # 操作对象 切换到打开的页面
        driver.switch_to.window(driver.window_handles[0])
    
    except:
        pass
```

#### 加载为 pandas 数据结构

```python
import pandas as pd

data = pd.DataFrame()
data["名称"] = titles
data["价格"] = prices
data["月销量"] = sale_counts
data["评论数"] = comment_counts

data
```
