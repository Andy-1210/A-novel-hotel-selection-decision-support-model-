# 首页数据爬取

```python
from selenium import webdriver
# 导入回车键的包
from selenium.webdriver.common.keys import Keys

from lxml import etree
import time
import pandas as pd

# 实例化一个浏览器对象(传入浏览器的驱动程序)
driver_path = r"E:\chromedriver\chromedriver.exe"    #我将chrome驱动安装的路径
driver = webdriver.Chrome(executable_path = driver_path)

# 让浏览器发起一个指定url对应的请求
driver.get("https://www.lagou.com")

# 关闭弹出的选项窗口
driver.find_element_by_xpath('//*[@id="cboxClose"]').click()

# 防止加载太快，页面数据还未显示
time.sleep(1)

# 找到搜索框 并输入关键字
driver.find_element_by_xpath('//*[@id="search_input"]').send_keys('python',Keys.ENTER)

# 关闭弹出的广告窗口
driver.find_element_by_xpath('/html/body/div[8]/div/div[2]').click() 

# 找到所有招聘信息的链接
position_links = driver.find_elements_by_xpath('//*[@id="s_position_list"]/ul/li/div[1]/div[1]/div[1]/a/h3')

titles = []
salarys = []
addrs = []
job_descs = []
for position_link in position_links:
    time.sleep(1)
    position_link.click()
    driver.switch_to.window(driver.window_handles[1])
    title = driver.find_element_by_xpath('/html/body/div[6]/div/div[1]/div/h1').text
    salary = driver.find_element_by_xpath('/html/body/div[6]/div/div[1]/dd/h3/span[1]').text
    addr = driver.find_element_by_xpath('/html/body/div[6]/div/div[1]/dd/h3/span[2]').text
    job_desc = driver.find_element_by_xpath('//*[@id="job_detail"]/dd[2]/div').text
    titles.append(title)
    salarys.append(salary)
    addrs.append(addr)
    job_descs.append(job_desc)
    driver.close()
    driver.switch_to.window(driver.window_handles[0])

data = pd.DataFrame()
data["名称"] = titles
data["薪水"] = salarys
data["地址"] = addrs
data["工作需求"] = job_descs
data


    
    
    
```

```python
data.to_csv("拉勾网python岗位信息.csv")
```

# 分页操作

```python
from selenium import webdriver
# 导入回车键的包
from selenium.webdriver.common.keys import Keys

from lxml import etree
import time
import pandas as pd

# 实例化一个浏览器对象(传入浏览器的驱动程序)
driver_path = r"E:\chromedriver\chromedriver.exe"    #我将chrome驱动安装的路径
driver = webdriver.Chrome(executable_path = driver_path)

# 让浏览器发起一个指定url对应的请求
driver.get("https://www.lagou.com")

# 关闭弹出的选项窗口
driver.find_element_by_xpath('//*[@id="cboxClose"]').click()

# 防止加载太快，页面数据还未显示
time.sleep(1)

# 找到搜索框 并输入关键字
driver.find_element_by_xpath('//*[@id="search_input"]').send_keys('python',Keys.ENTER)

# 关闭弹出的广告窗口
driver.find_element_by_xpath('/html/body/div[8]/div/div[2]').click() 

#page_source获取浏览器当前页面的页面源码数据
page_text = driver.page_source

# 先将第一页的源码数据存放到列表中，然后将后续获得的页面源码数据append到列表中
all_page_text = [page_text]
for i in range(6):
    # 进行下一页按钮的定位且对其进行点击
    next_tag = driver.find_element_by_xpath('//*[@id="s_position_list"]/div[2]/div/span[6]')
    next_tag.click()
    time.sleep(2)
    all_page_text.append(driver.page_source)

for page_text in all_page_text:
    # 找到所有招聘信息的链接
    position_links = driver.find_elements_by_class_name('position_link')

    titles = []
    salarys = []
    addrs = []
    detail_newss = []
    for position_link in position_links:
        time.sleep(1)
        position_link.find_element_by_tag_name('h3').click()
        driver.switch_to.window(driver.window_handles[1])
        title = driver.find_element_by_xpath('/html/body/div[6]/div/div[1]/div/h1').text
        salary = driver.find_element_by_xpath('/html/body/div[6]/div/div[1]/dd/h3/span[1]').text
        addr = driver.find_element_by_xpath('/html/body/div[6]/div/div[1]/dd/h3/span[2]').text
        detail_news = driver.find_element_by_xpath('//*[@id="job_detail"]/dd[2]/div').text
        titles.append(title)
        salarys.append(salary)
        addrs.append(addr)
        detail_newss.append(detail_news)
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

data = pd.DataFrame()
data["名称"] = titles
data["薪水"] = salarys
data["地址"] = addrs
data["工作需求"] = detail_newss
data
```
