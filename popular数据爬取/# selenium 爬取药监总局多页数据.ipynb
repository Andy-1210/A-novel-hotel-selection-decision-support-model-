```python
from selenium import webdriver
from lxml import etree
import time

#实例化一个浏览器对象(传入浏览器的驱动程序)
driver_path = r"E:\chromedriver\chromedriver.exe"    #我将chrome驱动安装的路径
driver = webdriver.Chrome(executable_path = driver_path)


#让浏览器发起一个指定url对应的请求
driver.get("http://125.35.6.84:81/xk/")

#page_source获取浏览器当前页面的页面源码数据
page_text = driver.page_source


# 先将第一页的源码数据存放到列表中，然后将后续获得的页面源码数据append到列表中
all_page_text = [page_text]
for i in range(6):
    # 进行下一页按钮的定位且对其进行点击
    next_tag = driver.find_element_by_xpath('//*[@id="pageIto_next"]')
    next_tag.click()
    time.sleep(2)
    all_page_text.append(driver.page_source)
    
for page_text in all_page_text:   
    tree = etree.HTML(page_text)
    li_list = tree.xpath('//*[@id="gzlist"]/li')
    for li in li_list:
        name = li.xpath('.//dl/@title')[0]
        print(name)
time.sleep(3)
driver.quit()  #访问结束，关闭浏览器
```
