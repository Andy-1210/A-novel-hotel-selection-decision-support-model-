```python
import pandas as pd
address="./马蜂窝数据提取.xlsx"
result = []
def excel_one_line_to_list():
    df = pd.read_excel(address, usecols=[0],
                       names=None)  # 读取第一列，不要列名
    df_li = df.values.tolist()

    for s_li in df_li:
        result.append(s_li[0])
    return result

if __name__ == '__main__':
    excel_one_line_to_list()

print(result.head())
```
